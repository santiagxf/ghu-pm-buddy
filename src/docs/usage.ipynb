Then, let's connect using the Azure AI Project client:

```python
!pip install azure-ai-projects --extra-index-url https://pkgs.dev.azure.com/azure-sdk/public/_packaging/azure-sdk-for-python/pypi/simple --pre --force-reinstall --quiet
```

```python
from azure.identity import DefaultAzureCredential
from azure.ai.projects import AIProjectClient

myEndpoint = "https://demo-fdp-swc.services.ai.azure.com/api/projects/demo-project-swc"

project_client = AIProjectClient(
    endpoint=myEndpoint,
    credential=DefaultAzureCredential(),
)
```

```python
myAgent = "pm-buddy-workflow"

agent = project_client.agents.get(agent_name=myAgent)
print(f"Retrieved agent: {agent.name}")
```

```python
openai_client = project_client.get_openai_client()
```

```python
response = openai_client.responses.create(
    input=[{"role": "user", "content": "You are assigned issue #1 in `santiagxf/travel-app`"}],
    extra_body={"agent": {"name": agent.name, "type": "agent_reference"}},
)

print(f"Response output: {response.output_text}")
```
